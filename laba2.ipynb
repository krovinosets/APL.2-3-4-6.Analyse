```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('Cleaned_ships_data.csv')
df.describe()
```

```python
sorted_df = df.sort_values('dwt')
fig = px.line(sorted_df, x="dwt", y="length")
fig.update_layout(
    title="Средняя максимальная загрузка кораблей от длины корабля",
    yaxis_title="Длина корабля",
    xaxis_title="Средняя максимальная загрузка"
)
fig.show()
```

```python
value_counts = df.value_counts("ship_name")
df_value_counts = pd.DataFrame(value_counts)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['ship_name', 'amount']

others = df_value_counts.loc[df_value_counts['amount'] < 11]
df_value_counts = df_value_counts.loc[df_value_counts['amount'] > 11]
df_value_counts.loc[len(df_value_counts)] = ["Others", others['amount'].sum()]

fig = px.histogram(df_value_counts, x='ship_name', y='amount')
fig.update_layout(
    title="Распределение количества кораблей по типам",
    xaxis_title="Тип корабля",
    yaxis_title="Количество"
)
fig.show()
```

```python
value_counts = df.value_counts("built_year")
df_value_counts = pd.DataFrame(value_counts)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['built_year', 'amount']

fig = px.histogram(df_value_counts, x='built_year', y='amount')
fig.update_layout(
    title="Гистограмма распределения количества кораблей по годам",
    xaxis_title="Год",
    yaxis_title="Количество"
)
fig.show()
```

```python
import plotly.graph_objects as go


fig = go.Figure(go.Histogram2d(
        x=df['built_year'],
        y=df['dwt'],
        colorscale="Viridis",
        nbinsy=20
    ))
fig.update_layout(
    title="2-D гистограмма распределения количества кораблей",
    yaxis_title="Максимальный вес груза",
    xaxis_title="Год постройки"
)
fig.show()
```

```python
df = pd.read_csv('Pubg_Stats.csv')
df.describe()
```

```python
sorted_df = df.sort_values('Wins')
fig = px.line(sorted_df, x="Wins", y="Top_10s")
fig.update_layout(
    title="Отношение частоты побед в отношении частоты попадания в топ 10",
    yaxis_title="Топ 10",
    xaxis_title="Побед"
)
fig.show()
```

```python
sorted_df = df.sort_values('Kills')
fig = px.line(sorted_df, x="Kills", y="Distance_Traveled")
fig.update_layout(
    title="Отношение частоты движения игрока в отношении количества убийств",
    yaxis_title="Пройденное расстояние",
    xaxis_title="Убийств"
)
fig.show()
```

```python
sorted_df = df.sort_values('Wins')
fig = px.line(sorted_df, x="Wins", y=["Revives", "Assists"])
fig.update_layout(
    title="Отношение командной игры в отношении частоты побед игрока",
    yaxis_title="Возрождения / Помощь",
    xaxis_title="Побед",
    legend_title="Командная активность"
)
fig.show()
```
